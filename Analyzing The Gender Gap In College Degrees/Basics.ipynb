This project will include the data for women in several majors during 1968-2011. The data will be plotted to find correlations between women and their selective majors.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

#create 3 categories for major
lib_arts_cats = ['Foreign Languages', 'English', 
'Communications and Journalism', 'Art and Performance', 
'Social Sciences and History']
other_cats = ['Health Professions', 
'Public Administration', 'Education', 'Agriculture',
'Business', 'Architecture']


fig = plt.figure(figsize=(18, 20))
#for STEM major, loop 6 times for 6 plots
for st in range(0,6):
    ax = fig.add_subplot(6,3,3*st+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[st]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[st]], c=cb_orange, label='Men', linewidth=3)
     #remove the spines
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[st])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    #set tick values
    ax.set_yticks([0,100])
    
    #adds a horizontal line at y=50 position
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    if st != 5:
        ax.tick_params(labelbottom='off')
    if st == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif st == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
#for liberal arts, loop 5 times for 5 plots 
for st in range(0,5):
    ax = fig.add_subplot(6,3,3*st+2)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[st]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[st]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[st])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    if st != 4:
        ax.tick_params(labelbottom='off')
    if st == 0:
        ax.text(2001, 82, 'Women')
        ax.text(2005, 20, 'Men')
#for other categories, loop 6 times for 6 plots           
for st in range(0, 6):
    ax = fig.add_subplot(6,3,3*st+3)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[st]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[st]], c=cb_orange, label='Men', linewidth=3)
   
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[st])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    if st != 5:
        ax.tick_params(labelbottom='off')
    if st == 0:
        ax.text(2001, 90, 'Women')
        ax.text(2005, 7, 'Men')
    elif st == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 32, 'Women') 
fig.savefig('gender_degrees.png')
    
plt.show()
```

STEM fields:
According to the charts, there is still a huge gender gap in engineering and Computer Science majors.   Women are increasingly pursuing physical sciences majors.

Liberal arts:
Most of the liberal arts majors are selected by women. There is no gender gap in history majors.

Other majors:
Majors such as agriculture, business, and architecture became more popular with women over the years while majors in health professions, and education are majority women.
